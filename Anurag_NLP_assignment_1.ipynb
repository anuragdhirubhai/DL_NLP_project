1. How was the DistilBERT model created?

DistilBERT is a small, fast, cheap and light Transformer model trained by distilling BERT base. It has 40% less parameters than bert-base-uncased and runs 60% faster while preserving over 95% of BERT's performances as measured on the GLUE language understanding benchmark.

The distillation process follows the teacher-student paradigm. BERT is the teacher and DistilBERT is the student. They are trained on the same language modeling objective. DistilBERT learns from the full output distribution (logits) of BERT, which is a softer and richer target than the originally masked words. This constitutes the main source of the transfer of knowledge from the teacher to the student during distillation.

```python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("distilbert-base-uncased")
```

```python
# Function to load and process dataset
def preprocess_function(examples):
    questions = [q.strip() for q in examples["question"]]
    inputs = tokenizer(
        questions,
        examples["context"],
        max_length=384,
        truncation="only_second",
        return_offsets_mapping=True,
        padding="max_length",
    )

    offset_mapping = inputs.pop("offset_mapping")
    answers = examples["answers"]
    start_positions = []
    end_positions = []

    for i, offset in enumerate(offset_mapping):
        answer = answers[i]
        start_char = answer["answer_start"][0]
        end_char = answer["answer_start"][0] + len(answer["text"][0])
        sequence_ids = inputs.sequence_ids(i)

        # Find the start and end of the context
        idx = 0
        while sequence_ids[idx] != 1:
            idx += 1
        context_start = idx
        while sequence_ids[idx] == 1:
            idx += 1
        context_end = idx - 1

        # If the answer is not fully inside the context, label it (0, 0)
        if offset[context_start][0] > end_char or offset[context_end][1] < start_char:
            start_positions.append(0)
            end_positions.append(0)
        else:
            # Otherwise it's the start and end token positions
            idx = context_start
            while idx <= context_end and offset[idx][0] <= start_char:
                idx += 1
            start_positions.append(idx - 1)

            idx = context_end
            while idx >= context_start and offset[idx][1] >= end_char:
                idx -= 1
            end_positions.append(idx + 1)

    inputs["start_positions"] = start_positions
    inputs["end_positions"] = end_positions
    return inputs
```

```python
tokenized_squad_train = dataset_train.map(
    preprocess_function, 
    batched=True, 
    remove_columns=squad["train"].column_names
    )
```

```python
tokenized_squad_validation = dataset_validation.map(
    preprocess_function, 
    batched=True, 
    remove_columns=squad["train"].column_names
    )
```

```python
from transformers import DefaultDataCollator

data_col = DefaultDataCollator(return_tensors="tf")
```

```python
from transformers import create_optimizer

batch_size = 16
num_epochs = 2
total_train_steps = (len(tokenized_squad_train) // batch_size) * num_epochs
optimizer, schedule = create_optimizer(
    init_lr=2e-5,
    num_warmup_steps=0,
    num_train_steps=total_train_steps,
)
```

```python
from transformers import TFAutoModelForQuestionAnswering

model = TFAutoModelForQuestionAnswering.from_pretrained("distilbert-base-uncased")
```

```python
#Convert your datasets to the `tf.data.Dataset`

tf_train_set = model.prepare_tf_dataset(
    tokenized_squad_train,
    shuffle=True,
    batch_size = 32,
    collate_fn = data_col,
)

tf_validation_set = model.prepare_tf_dataset(
    tokenized_squad_validation,
    shuffle = False,
    batch_size = 32,
    collate_fn = data_col,
)
```

```python
model.compile(optimizer=optimizer)
```

```python
#setting a way to push the model to hub

from transformers.keras_callbacks import PushToHubCallback

callback = PushToHubCallback(
    output_dir="my_awesome_qa_model",
    tokenizer=tokenizer,
)
```

```python
model.fit(x=tf_train_set, validation_data=tf_validation_set, epochs=3, callbacks=[callback])

```
